```python
# IMPORTS
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# PROBLEM 1A

z = np.linspace(-3,3,1000)
activationFunction_1 = 1/(1+np.exp(-z)) 
activationFunction_2 = np.tanh(z)
activationFunction_3 = np.maximum(0,z)

plt.figure(figsize=(8,8))
plt.plot(z, activationFunction_1, color='blue', label=r'$\phi_1(z)$ [Sigmoid]')
plt.plot(z, activationFunction_2, color='red', label=r'$\phi_2(z)$ [Tanh]')
plt.plot(z, activationFunction_3, color='green', label=r'$\phi_3(z)$ [ReLU]')

plt.xlabel('Activation Input (z)')
plt.ylabel(r'Activation Output')
plt.title('Activation Functions')
plt.legend(loc='best')
plt.grid('True')
```

```python
# Problem 3E
# Make changes to the dataframe: add column, delete column, use multiple columns to create a new column, rearrange columns

print('Add Bedroom Ratio Column')
df = pd.read_csv('housing.csv')
df.sort_values('median_house_value')
df['bedroom_ratio'] = round(df['total_bedrooms'] / df['total_rooms'], 4)
print(df.head(n=5))
print('\n\n')

print('Delete Total Bedrooms Column')
df = df.drop(columns=['total_bedrooms'])
print(df.head(n=5))
print('\n\n')

print('Use ratio of Median House Value and Median Income columns to create Affordability Ratio Column')
df['affordability_ratio'] = round(df['median_house_value'] / df['median_income'], 1)    
print('\n\n')

print('Rearrange: Move Bedroom Ratio after Total Bedroom and Affordability Ratio after Median House Value')
cols = list(df.columns)
df = df[cols[0:4] + [cols[-2]] + cols[4:8] + [cols[-1]] + [cols[-3]]]
print(df.head(n=5))
print('\n\n')
```

```python
# Problem 3F
# Filter data

print('Filter House less than $250,000 and Near Ocean or Near Bay')
new_df = df.loc[ ((df['ocean_proximity'] == 'NEAR BAY') | (df['ocean_proximity'] == 'NEAR OCEAN')) & (df['median_house_value'] < 250000)]
new_df
```

```python
# Problem 3G
# Reset index
new_df = new_df.reset_index()
new_df
```

```python
# Problem 3H
# Aggregate Statistics using GroupBy
print('Mean Values Grouped by Ocean Proximity, Sorted by Median House Value')
df.groupby(['ocean_proximity']).mean().sort_values('median_house_value')
```
