```python
# Question 1A
import numpy as np

# The two matrices to multiply
matrix1 = np.array([(1, 2, 4), (3, 4, 2)])
matrix2 = np.array([[4], [2]])

# Answer with matrix multiplication
answer = matrix1 @ matrix2
print(answer)
```

Operation 1A is invalid because matrix multiplication requires the number of columns in the first array equal the number of rows in the second array. The second array has only 2 rows, not 3 rows, so matrix multiplication is not possible. 

```python
# Question 1B

# The two matrices to multiply
matrix1 = np.array([(1, 2, 4), (3, 4, 2)])
matrix2 = np.array([[4], [5], [2]])

# Answer with matrix multiplication
answer = matrix1 @ matrix2
print(answer)
```

```python
# Question 1C

# The two matrices to multiply
matrix1 = np.array([(1, 2, 4), (3, 4, 2)])
matrix2 = np.array([(4, 2)])

# Answer with matrix multiplication
answer = matrix1 @ matrix2
print(answer)
```

Operation 1C is invalid because matrix multiplication requires the number of columns in the first array equal the number of rows in the second array. The second array has only 1 row, not 3 rows, so matrix multiplication is not possible.

```python
# Question 1D

# The two matrices to multiply
# The first matrix is a transpose of another matrix
matrix1_untransposed = np.array([[6], [2], [4]])
matrix1 = np.transpose(matrix1_untransposed)
matrix2 = np.array([[5], [2], [1]])

# Answer with matrix multiplication
answer = matrix1 @ matrix2
print(answer)
```

```python
# Question 2

# Solve system of linear equations with Ax = b OR x = (A_inverse)(b)
# In this case, x is a vector representing x, y, z in the original system of linear equations
A = np.array([(2, 3, 5), (1, 1, 1), (1, -2, -3)])
b = np.array([[17], [6], [-10]])
A_inverse = np.linalg.inv(A)

# Answer with x = (A_inverse)(b)
x = A_inverse @ b
print(x)
```

Based on the solution, x = 0.6, y = 5.6, and z = -0.2.

```python
# Question 3
n = 1000

# Part A: Normalize such that ||x||2 = 1
# numpy.random.randn is used to generate n vectors in an n x 2 array with a Gaussian distribution
# Each row represents a vector. Each column represents the 1st or 2nd element of all vectors.
# The L2 norm for each row vector is calculated and used to scale all vectors such that ||x||2 = 1
# Due to a quirk of python, the scaling vector needed to be explicitly reshaped
xA_unscaled = np.random.randn(n,2)
xA_col1 = xA_unscaled[:, 0]
xA_col2 = xA_unscaled[:, 1]
xA_sum_squares = np.square(xA_col1) + np.square(xA_col2)
xA_scalings = np.sqrt(xA_sum_squares).reshape(n, 1)
xA = xA_unscaled / xA_scalings

# Part B: Normalize such that ||x||1 = 1
# Similar to part A, but normalize by L1 norm that is simply the sum of the absolute value of the elements. 
xB_unscaled = np.random.randn(n,2)
xB_col1 = xB_unscaled[:, 0]
xB_col2 = xB_unscaled[:, 1]
xB_scalings = (np.abs(xB_col1) + np.abs(xB_col2)).reshape(n, 1)
xB = xB_unscaled / xB_scalings

# Part C: Normalize such that ||x||1 = 2
# Similar to part A, but multiply by 2
xC_unscaled = np.random.randn(n,2)
xC_col1 = xC_unscaled[:, 0]
xC_col2 = xC_unscaled[:, 1]
xC_sum_squares = np.square(xC_col1) + np.square(xC_col2)
xC_scalings = np.sqrt(xC_sum_squares).reshape(n, 1)
xC = 2 * xC_unscaled / xC_scalings

# Part D: Normalize such that ||x||1 = 2
# Similar to part B, but multiply by 2
xD_unscaled = np.random.randn(n,2)
xD_col1 = xD_unscaled[:, 0]
xD_col2 = xD_unscaled[:, 1]
xD_scalings = (np.abs(xD_col1) + np.abs(xD_col2)).reshape(n, 1)
xD = 2 * xD_unscaled / xD_scalings

# Plot Script
# The grid size was changed to an 8 x 8 square to prevent distortions. 
plt.figure(figsize=(8,8)) 
plt.scatter(xA[:, 0], xA[:, 1], color='blue', label=r'$||x||_2 = 1$')
plt.scatter(xB[:, 0], xB[:, 1], color='red', label=r'$||x||_1 = 1$')
plt.scatter(xC[:, 0], xC[:, 1], color='green', label=r'$||x||_2 = 2$')
plt.scatter(xD[:, 0], xD[:, 1], color='orange', label=r'$||x||_1 = 2$')

plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title('Random n vectors x of dimension 2 x 1, n = 1000')
plt.legend(loc='upper right')
plt.grid('True')
```

Based on this plot, it's apparent that all points with the same L2 norm lie on a circle with radius equal to the norm and center at the origin. 

Similarly, all points with the same L1 norm lie on a 45 degree rotated square with sides equal to the L1 norm and center at the origin. 

```python
# Question 4
n = 100
variance = 1
a = 2
b = 3

# Part A: Create Gaussian random vector of dimenson = n x 1, mean = 0, and variance = 1
# numpy.random.normal is used to generate a Gaussian random vector with specified mean and standard deviation
# Standard deviation is the square root of variance
mean = 0
stdev = np.sqrt(variance)
X = np.random.normal(mean, stdev, size=(n, 1))

# Part B: Create vector Y = aX + b
Y = a*X + b

# Part C: Create Y_noisy = Y + Z, where Z is Gaussian random variable with mean = zero, variance = 1
Z = np.random.normal(mean, stdev, size=(n, 1))
Y_noisy = Y + Z

# Part D: Find correlation coefficientg between X and Y with np.corrcoef(X, Y, rowvar=False)
corr_matrix = np.corrcoef(X, Y_noisy, rowvar=False)
corr = corr_matrix[1, 0]
print('Correlation coefficient between X and Y_noisy is ' + str(corr))

# Part E: Scatter Plot of Y_noisy vs. X
plt.figure()
plt.scatter(X, Y_noisy, color='blue')
plt.xlabel(r'$X$')
plt.ylabel(r'$Y_noisy$')
plt.title(fr'Scatter Plot $\mu=0, \sigma^2=1$, n=100, a=2, b=3, r={corr:.4f}')
plt.grid('True')

# Part F.1: Change n and repeat A-E
n = 10000
variance = 1
a = 2
b = 3

mean = 0
stdev = np.sqrt(variance)
X = np.random.normal(mean, stdev, size=(n, 1))

Y = a*X + b
Z = np.random.normal(mean, stdev, size=(n, 1))
Y_noisy = Y + Z

corr_matrix = np.corrcoef(X, Y_noisy, rowvar=False)
corr = corr_matrix[1, 0]
print('Changing n = 10000, correlation coefficient between X and Y_noisy is ' + str(corr))

plt.figure()
plt.scatter(X, Y_noisy, color='blue')
plt.xlabel(r'$X$')
plt.ylabel(r'$Y_noisy$')
plt.title(fr'Scatter Plot $\mu=0, \sigma^2=1$, n=10000, a=2, b=3, r={corr:.4f}')
plt.grid('True')

# Part F.2: Change variance and repeat A-E
n = 100
variance = 10000
a = 2
b = 3

mean = 0
stdev = np.sqrt(variance)
X = np.random.normal(mean, stdev, size=(n, 1))

Y = a*X + b
Z = np.random.normal(mean, stdev, size=(n, 1))
Y_noisy = Y + Z

corr_matrix = np.corrcoef(X, Y_noisy, rowvar=False)
corr = corr_matrix[1, 0]
print('Changing variance = 10000, correlation coefficient between X and Y_noisy is ' + str(corr))

plt.figure()
plt.scatter(X, Y_noisy, color='blue')
plt.xlabel(r'$X$')
plt.ylabel(r'$Y_noisy$')
plt.title(fr'Scatter Plot $\mu=0, \sigma^2=10000$, n=100, a=2, b=3, r={corr:.4f}')
plt.grid('True')

# Part F.3: Change a and repeat A-E
n = 100
variance = 1
a = 200
b = 3

mean = 0
stdev = np.sqrt(variance)
X = np.random.normal(mean, stdev, size=(n, 1))

Y = a*X + b
Z = np.random.normal(mean, stdev, size=(n, 1))
Y_noisy = Y + Z

corr_matrix = np.corrcoef(X, Y_noisy, rowvar=False)
corr = corr_matrix[1, 0]
print('Changing a = 200, correlation coefficient between X and Y_noisy is ' + str(corr))

plt.figure()
plt.scatter(X, Y_noisy, color='blue')
plt.xlabel(r'$X$')
plt.ylabel(r'$Y_noisy$')
plt.title(fr'Scatter Plot $\mu=0, \sigma^2=1$, n=100, a=200, b=3, r={corr:.4f}')
plt.grid('True')

# Part F.4: Change b and repeat A-E
n = 100
variance = 1
a = 2
b = 300

mean = 0
stdev = np.sqrt(variance)
X = np.random.normal(mean, stdev, size=(n, 1))

Y = a*X + b
Z = np.random.normal(mean, stdev, size=(n, 1))
Y_noisy = Y + Z

corr_matrix = np.corrcoef(X, Y_noisy, rowvar=False)
corr = corr_matrix[1, 0]
print('Changing b = 300, correlation coefficient between X and Y_noisy is ' + str(corr))

plt.figure()
plt.scatter(X, Y_noisy, color='blue')
plt.xlabel(r'$X$')
plt.ylabel(r'$Y_noisy$')
plt.title(fr'Scatter Plot $\mu=0, \sigma^2=1$, n=100, a=2, b=300, r={corr:.4f}')
plt.grid('True')
```

The script was ran several times to make the appropriate general observations. 

For the original data:
- In Part D, the correlation coefficient was generally around 0.88 - 0.92
- In Part E, the scatter plot was roughly distributed around the line with y-intercept of -3 and slope of 2, reflecting the linear mapping of Y = aX + b. 

The following was noted with changes in sample size:
- As n decreased, the correlation coefficient was more variable across different simulations.
- As n increased, the correlation coefficient settled more tightly around 0.90. This is because the larger sample size better captured the variability that was baked into Y_noisy. 
- Additionally, larger n resulted in a wider range of X values, since more samples increased the chance of observing less likely values in the Gaussian distribution.

The following was noted with changes in variance:
- As variance increased, there was a wider range of X values.
- There was also a wider range of Y_noisy values resulting from both the variance in X, and the variance added from Z.
- Due to this additonal variance, the correlation coefficient was generally lower.

The following was noted with changes in a:
- As a increased, the values of Y_noisy increased.
- This improved the correlation coefficient because Y was much much larger than the noise generated in Z, resulting in Y_noisy being closer to the original Y.

The following was noted with changes in b:
- The Y values were offset, but the general linear relationship, variability and correlation coefficient didn't appear to change.
